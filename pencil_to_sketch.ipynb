```python
import cv2
from IPython.display import Image
from google.colab.patches import cv2_imshow
```

**Trying To read image**

```python
image=cv2.imread("sample.jpg")
cv2_imshow(image)
```

```python
gray_image = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
cv2_imshow(gray_image)
```

```python
inv_image=255-gray_image
cv2_imshow(inv_image)
```

```python
blr=cv2.GaussianBlur(inv_image,(21,21),0)
cv2_imshow(blr)
```

```python
inv_blr=255-blr
cv2_imshow(inv_blr)
```

**final sketch image **

```python
p_sketch=cv2.divide(gray_image,inv_blr,scale=267)
cv2_imshow(p_sketch)
```
